```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/titanic-train/train.csv')
print(df)
```

```python
cols = ['Name', 'Ticket', 'Cabin']
df = df.drop(cols, axis=1)
```

```python
dummies = []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
    dummies.append(pd.get_dummies(df[col]))
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
```

```python
df.head(10)
```

```python
df = pd.concat((df,titanic_dummies), axis=1)
```

```python
df = df.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
```

```python
df['Age'] = df['Age'].interpolate()
```

```python
print(df)
```

```python
df.head(10)
```

```python
# Y_train = df["Survived"]
# X_train = df.drop(["Survived"], axis = 1 )

# X_test = a
# X_train.shape, y_train.shape, X_test.shape
X = df.drop("Survived",axis=1)
y = df["Survived"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
accuracies = []
```

```python
LR = LogisticRegression()

LR.fit(X_train, y_train)
y_pred_lr = LR.predict(X_test)
LR_score1 = LR.score(X_train, y_train)
print("LR Accuracy  score = {:.2f}".format(LR_score1*100))
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2,random_state = 6)
X_train = df.drop(['Survived'], axis=1)
y_train = df['Survived']
logmodel = LogisticRegression(random_state = 100)
logmodel.fit(X_train, y_train)

print(logmodel.score(X_test,y_test))
pp = logmodel.predict(X_test)

LR_score1=logmodel.score(X_test,y_test)

```

```python
X = df.drop("Survived",axis=1)
y = df["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = .3,
                                                    random_state = 5,
                                                   stratify = y)

rf = RandomForestClassifier(n_estimators=100, random_state=0)

rf.fit(X_train, y_train)

rf_score = rf.score(X_test, y_test)
print("RF Accuracy  score = {:.2f}".format(rf_score*100))
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
# RFC Parameters tunning 
K_fold = StratifiedKFold(n_splits=10)
RFC = RandomForestClassifier()

## Search grid for optimal parameters
rf_param_grid = {"max_depth": [None],
              "min_samples_split": [2, 6,10, 20],
              "min_samples_leaf": [1, 4,8 ,16],
              "n_estimators" :[100,200,300,400,500],
              "criterion": ["gini"]}


gsRFC = GridSearchCV(RFC, param_grid = rf_param_grid, cv=K_fold,
                     scoring="accuracy", n_jobs= 2, verbose = 1)

gsRFC.fit(X_train,y_train)
RFC_best = gsRFC.best_estimator_

# Best score
gsRFC.best_score_

```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
svm_score = svm.score(X_test, y_test)
accuracies.append(svm_score)
print("SVM Accuracy  score = {:.2f}".format(svm_score*100))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear',probability=True, C=4, gamma=0.1, max_iter=-1,random_state=10)
svm.fit(X_train, y_train)
svm_score1 = svm.score(X_test, y_test)
accuracies.append(svm_score)
print("SVM Accuracy  score = {:.2f}".format(svm_score1*100))
```
